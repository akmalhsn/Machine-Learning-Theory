# DBSCAN

- DBSCAN merupakan clustering berbasis kepadatan yaang mengacu pada jumlah titik yang terletak di suatu area, dimana semakin dekat suatu titik dengan titik lainnya maka akan semakin besar kepadatannya. 
- DBSCAN umum digunakan pada data yang mengandung banyak noise, karena data noise akan dianggap kosong.
- DBSCAN tidak mengharuskan pengguna untuk memasukan jumlah cluster seperti K-Means

DBSCAN memerlukan dua parameter input untuk melakukan clustering yaitu,
- epsilon, jarak maksimum antara titik x terhadap sekelilingnya
- minimum points, jumlah minimum anggota yang ada disekitar epsilon untuk dianggap sebagai cluster 

Pembentukan Cluster DBSCAN dipengaruhi oleh tiga jenis titik data yaitu,
- titik noise/anomali, titik yang tidak memiliki cukup tetangga dalam epsilon
- titik border/batas, titik yang memiliki tetangga dalam epsilon tetapi jumlahnya kurang dari minimum points
- titik inti, jumlah tetangga lebih dari atau sama dengan minimum points



Berikut Tahapan DBSCAN
1. Menentukan epsilon dan minPts
2. memeilih titik `p` secara acak
3. menghitung jarak antara titik `p` terhadap semua data menggunakan euclidian distance
4. mengambil semua titik yang dapat dijangkau kepadatan dari `p` berdasarkan minPTS dan epsilon
5. jika `p` adalah titik inti maka sebuah cluster terbentuk
6. jika `p` adalah titik batas dan tidak ada titik yang dapat menjangkau kepadatan dari `p` maka DBSCAN melanjutkan ke titik berikutnya
7. ulangi langkah 3 sampai 6 hingga semua poin selesai diproses

## 1. Import Library

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

## 2. Membaca Dataset

Dataset yang digunakan berasal dari library sklearn

```python
# membuat klasifikasi data 2D sebanyak 500 
# noise sebanyak 10%, noise dapat membuat dataset lebih realistis
X, y = make_moons(n_samples=500, noise=0.1)
```

## 3. Visualisasi Dataset

membuat scatter plot dari dataset yang dihasilkan dengan make_moons, di mana setiap titik pada plot diwarnai berdasarkan kelasnya (0 atau 1), dan sumbu x dan y diberi label "$x_1$" dan "$x_2$".

```python
plt.scatter(X[:, 0], X[:, 1], c=y, label=y)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
```

## 4. Pemodelan DBSCAN

```python
#Using Default parameters
dbscan_cluster1 = DBSCAN()
dbscan_cluster1.fit(X)
plt.scatter(X[:, 0], 
            X[:, 1], 
            c=dbscan_cluster1.labels_, 
            label=y)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
```

Dengan merubah parameter seperti eps dan min_samples dapat mengendalikan bagaimana DBSCAN mengenali cluster dalam data. Perubahan nilai parameter dapat mempengaruhi ukuran dan bentuk dari cluster yang dihasilkan oleh algoritma. 

```python
#Changing the parameters
dbscan_cluster2 = DBSCAN(eps=0.2, min_samples=5)
dbscan_cluster2.fit(X)
plt.scatter(X[:, 0], 
            X[:, 1], 
            c=dbscan_cluster2.labels_, 
            label=y)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
```

```python
#another parameter
dbscan_cluster3 = DBSCAN(eps=0.2, min_samples=15)
dbscan_cluster3.fit(X)
plt.scatter(X[:, 0], 
            X[:, 1], 
            c=dbscan_cluster3.labels_, 
            label=y)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
```

```python
#another parameter
dbscan_cluster4 = DBSCAN(eps=0.2, min_samples=25)
dbscan_cluster4.fit(X)
plt.scatter(X[:, 0], 
            X[:, 1], 
            c=dbscan_cluster4.labels_, 
            label=y)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
```

## Kelebihan dan Kekurangan DBSCAN

### Kelebihan
- mampu melakukan identifikasi noise saat membuat cluster 
- mampu menemukan ukuran cluster yang baik

### Kekurangan
- sulit pada data berdimensi tinggi
- gagal dalam kasus density cluster
